```python
import rdflib
import os
from rdflib import URIRef, BNode, Literal
from rdflib.namespace import Namespace, NamespaceManager, RDF, FOAF, RDFS, OWL
from rdflib.plugins.sparql import prepareQuery
import glob
import urllib.parse as urlparse
import urllib.request as urllib

namespace_manager = NamespaceManager(rdflib.Graph())
namespace_manager.bind('foaf', FOAF, override=False)
namespace_manager.bind('rdfs', RDFS, override=False)
namespace_manager.bind('sesame', Namespace('http://www.openrdf.org/schema/sesame#'), override = False)
namespace_manager.bind('ra', Namespace('http://bio2rdf.org/ra.challenge_vocabulary:'), override = False)
namespace_manager.bind('ncbigene', Namespace('http://bio2rdf.org/ncbigene_vocabulary:'), override = False)
SAMWALD = Namespace('tag:samwald.info:')
namespace_manager.bind('samwald', SAMWALD, override = False)

endpoint = 'http://localhost:8080/openrdf-sesame/repositories/synapse-challenge-2' # this is an RDF store (no inferencing)
g = rdflib.ConjunctiveGraph('SPARQLUpdateStore', identifier='http://samwald.info/ont/') 
g.namespace_manager = namespace_manager

g.open((endpoint, endpoint + '/statements')) # endpoint, SPARQL Update endpoint
```

```python
from IPython.display import display

class ListTable(list):
    """ Overridden list class which takes a 2-dimensional list of 
        the form [[1,2,3],[4,5,6]], and renders an HTML Table in 
        IPython Notebook. """
    
    def _repr_html_(self):
        html = ["<table style='font-size:10px'>"]
        for row in self:
            html.append("<tr>")
            
            for col in row:
                html.append("<td>{0}</td>".format(col))
            
            html.append("</tr>")
        html.append("</table>")
        return ''.join(html)
```

## Genetic data in detail

#### Create data on polymorphism variant counts in graph <tag:samwald.info:polymorphism_variant_counts>

```python
q = '''
PREFIX ra: <http://bio2rdf.org/ra.challenge_vocabulary:>

INSERT { 
    GRAPH <tag:samwald.info:polymorphism_variant_counts> { 
        ?polymorphism_variant ra:count ?count .
    }
}
WHERE {
    SELECT ?polymorphism_variant (COUNT(DISTINCT ?person) AS ?count) WHERE {
        ?person ra:has-one | ra:has-two ?polymorphism_variant .
    } 
    GROUP BY ?polymorphism_variant
}
'''

g.update(q)
```

#### Let's see the major and minor variants for each polymorphism (together with counts of how many patients have at least one such variant)

```python
q = '''\
SELECT DISTINCT ?polymorphism_variant_major ?count_major ?polymorphism_variant_minor ?count_minor
WHERE {
    ?polymorphism rdfs:subClassOf ra:Polymorphism .
    ?polymorphism_variant_major a ?polymorphism ;
        ra:count ?count_major .
    ?polymorphism_variant_minor a ?polymorphism ;
        ra:count ?count_minor .
        
    FILTER(!sameTerm(?polymorphism_variant_major, ?polymorphism_variant_minor) && ?count_major > ?count_minor)
}
ORDER BY DESC(?count_major)
'''

table = ListTable()
table.append(['polymorphism_variant_major', 'count_major', 'polymorphism_variant_minor', 'count_minor'])
for row in g.query(q):
    table.append([row.polymorphism_variant_major, row.count_major, row.polymorphism_variant_minor, row.count_minor])
    count += 1
table
```

#### This is cool, let's also reflect this in the RDF by inferring ra:major_variant and ra:minor_variant types and adding those to the graph

```python
q = '''

INSERT { 
    GRAPH <tag:samwald.info:polymorphism_variant_counts> {
        ?polymorphism_variant_major a ra:major_variant .
        ?polymorphism_variant_minor a ra:minor_variant .
    }
} WHERE {
SELECT DISTINCT ?polymorphism_variant_major ?count_major ?polymorphism_variant_minor ?count_minor
WHERE {
    ?polymorphism rdfs:subClassOf ra:Polymorphism .
    ?polymorphism_variant_major a ?polymorphism ;
        ra:count ?count_major .
    ?polymorphism_variant_minor a ?polymorphism ;
        ra:count ?count_minor .
        
    FILTER(!sameTerm(?polymorphism_variant_major, ?polymorphism_variant_minor) && ?count_major > ?count_minor)
}
}'''

g.update(q)
```

#### Number of patients with homozygous minor variants

```python
q = '''\
SELECT ?polymorphism_variant (COUNT(DISTINCT ?person) AS ?polymorphism_count) WHERE {
        ?person ra:has-two ?polymorphism_variant .
        ?polymorphism_variant a ra:minor_variant .
    } 
GROUP BY ?polymorphism_variant
ORDER BY DESC(?polymorphism_count)
'''

table = ListTable()
table.append(['polymorphism_variant', 'polymorphism_count'])
for row in g.query(q):
    table.append([row.polymorphism_variant, row.polymorphism_count])
table
```

#### For each minor variant, show the number and percentage of responders vs. non-responders that are homozygous for that variant

```python
q = '''\
SELECT ?polymorphism_variant (COUNT(DISTINCT ?responder) AS ?responder_count)(COUNT(DISTINCT ?non_responder) AS ?non_responder_count) WHERE {
?polymorphism_variant a ra:minor_variant .
    {
        ?responder ra:has-two ?polymorphism_variant .
        ?responder ra:non-responder false . 
    }
    UNION 
    {
        ?non_responder ra:has-two ?polymorphism_variant .
        ?non_responder ra:non-responder true .
    }
} 
GROUP BY ?polymorphism_variant
ORDER BY DESC(?non_responder_count)
'''

table = ListTable()
table.append(['polymorphism_variant', 'number of responders with this variant', '(in percent)', 'number of non-responders with this variant', '(in percent)', 'Difference of percentage'])
for row in g.query(q):
    table.append([row.polymorphism_variant, 
          row.responder_count, 
          '{0:.0%}'.format(int(row.responder_count) / number_of_responders), 
          row.non_responder_count, 
          '{0:.0%}'.format(int(row.non_responder_count) / number_of_non_responders),
          '{0:.0%}'.format(int(row.non_responder_count) / number_of_non_responders - int(row.responder_count) / number_of_responders)])
table
```

#### This is quite inconclusive. In total per patient, do non-responders have more 'minor variants' than responders? (again we are only comparing homozygous cases, based on the assumption that this is where we could see the strongest effect)

```python
q = '''\
SELECT (COUNT(DISTINCT ?person) as ?person_count) (COUNT(?polymorphism_variant) as ?number_of_total_homozygous_minor_variant_genotypes) ?non_responder WHERE {
    ?person ra:non-responder ?non_responder ;
        ra:has-two ?polymorphism_variant .
    ?polymorphism_variant a ra:minor_variant .
}
GROUP BY ?non_responder
'''

for row in g.query(q):
    print('There are', row.person_count, 'persons with non_responder status', 
          row.non_responder, 'with a population-wide total of', row.number_of_total_homozygous_minor_variant_genotypes,
          'homozygous minor variant genotypes --', '{0:.5f}'.format(int(row.number_of_total_homozygous_minor_variant_genotypes) / int(row.person_count)), 'per person')
```

#### Let's see what kinds of GO annotations (processes and functions) the genes we are observing have.

```python
q = '''\
SELECT ?go_term_label (COUNT(DISTINCT ?snp) as ?snp_count) (GROUP_CONCAT(DISTINCT ?gene_id) as ?gene_id_concat) {
    ?snp a <http://bio2rdf.org/dbsnp_vocabulary:snp> ;
        <http://bio2rdf.org/dbsnp_vocabulary:maps-to> ?gene .
    ?gene a <http://bio2rdf.org/ncbigene_vocabulary:Resource> ;
        <http://purl.org/dc/terms/identifier> ?gene_id ;
        ncbigene:process | ncbigene:function ?go_term .
    ?go_term rdfs:label ?go_term_label .
}
GROUP BY ?go_term_label
ORDER BY DESC(?snp_count)
'''

count = 0
table = ListTable()
table.append(['GO term', 'Number of SNPs associated with this term', 'Corresponding genes'])
for row in g.query(q):
    table.append([row.go_term_label, row.snp_count, row.gene_id_concat])
    count += 1

print(count, 'GO terms in total')
display(table)
```

#### Let's also include all the superclasses of the GO terms used as annotations for the SNPs

```python
# note: this only works on the simple rdfs:subClassOf hierarchy, it will not take care of complex OWL axioms in the GO hierarchy
q = '''\
SELECT ?go_term_label (COUNT(DISTINCT ?snp) as ?snp_count) {
    ?snp a <http://bio2rdf.org/dbsnp_vocabulary:snp> ;
        <http://bio2rdf.org/dbsnp_vocabulary:maps-to> ?gene .
    ?gene a <http://bio2rdf.org/ncbigene_vocabulary:Resource> ;
        ncbigene:process/rdfs:subClassOf* | ncbigene:function/rdfs:subClassOf* ?go_term .
    ?go_term rdfs:label ?go_term_label .
}
GROUP BY ?go_term_label
ORDER BY DESC(?snp_count)
'''

set_of_go_terms_associated_with_SNPs = set()
table = ListTable()
table.append(['GO term (including inferred)', 'SNPs associated with this GO term'])
count = 0 
for row in g.query(q):
    table.append([row.go_term_label, row.snp_count])
    set_of_go_terms_associated_with_SNPs.add(str(row.go_term_label))
    count += 1
print(count, 'GO terms (including inferred) in total')
display(table)
```

#### Now, let's see the GO annotations the targets of the drugs in the study have. We do this via Drugbank and Uniprot.

```python
q = '''\
PREFIX goa: <http://bio2rdf.org/goa_vocabulary:>

SELECT ?go_term_label 
       (GROUP_CONCAT(DISTINCT ?drug_label; SEPARATOR = "; ") as ?drug_label_concat) 
       (GROUP_CONCAT(DISTINCT ?protein_label; SEPARATOR = "; ") as ?protein_label_concat) 
WHERE {
    ?drug a <http://bio2rdf.org/ra.challenge:Drug> ;
        rdfs:label ?drug_label ;
        <http://bio2rdf.org/drugbank_vocabulary:target> ?protein .
    ?protein a <http://bio2rdf.org/uniprot_vocabulary:Resource> ;
        rdfs:label ?protein_label ;
        goa:process | goa:function ?go_term .
    ?go_term rdfs:label ?go_term_label .
}
GROUP BY ?go_term_label
ORDER BY DESC(COUNT(DISTINCT ?drug))
'''

set_of_go_terms_associated_with_drug_targets = set()
for row in g.query(q):
    print('GO term:\n', row.go_term_label)
    print('Drugs associated with GO term:\n', row.drug_label_concat)
    print('Through these drug targets:\n', row.protein_label_concat)
    print('\n---\n')
    set_of_go_terms_associated_with_drug_targets.add(str(row.go_term_label))
```

## Clinical parameters

### Responders and non-responders for each drug regime

```python
q = '''\
SELECT ?label (COUNT(DISTINCT ?responder) AS ?responder_count)(COUNT(DISTINCT ?non_responder) AS ?non_responder_count) WHERE {
    {
        ?responder ra:drug ?drug .
        ?responder ra:non-responder false . 
        ?drug rdfs:label ?label .
    }
    UNION 
    {
        ?non_responder ra:drug ?drug .
        ?non_responder ra:non-responder true .
        ?drug rdfs:label ?label .
    }
} 
GROUP BY ?label
ORDER BY DESC(?non_responder_count)
'''

for row in g.query(q):
    total_patient_number_for_this_drug = int(row.responder_count) + int(row.non_responder_count)
    print(total_patient_number_for_this_drug, "patients in total took", row.label, 
          "of which", row.responder_count, 
          '({0:.0%})'.format(int(row.responder_count) / total_patient_number_for_this_drug), 'were responders and',
          row.non_responder_count, 
          '({0:.0%})'.format(int(row.non_responder_count) / total_patient_number_for_this_drug), 'were non-responders')
```

#### Seems like patients on adalimumab had considerably fewer non-responders. Let's also take methotrexate co-therapy into account.

```python
q = '''\
SELECT ?label 
       (COUNT(DISTINCT ?responder) AS ?responder_count)
       (COUNT(DISTINCT ?non_responder) AS ?non_responder_count) 
       ?methotrexate_cotherapy
WHERE {
    {
        ?responder ra:drug ?drug .
        ?responder ra:non-responder false . 
        ?drug rdfs:label ?label .
        ?responder ra:methotrexate-cotherapy ?methotrexate_cotherapy .
    }
    UNION 
    {
        ?non_responder ra:drug ?drug .
        ?non_responder ra:non-responder true .
        ?drug rdfs:label ?label .
        ?non_responder ra:methotrexate-cotherapy ?methotrexate_cotherapy .
    }
} 
GROUP BY ?label ?methotrexate_cotherapy
ORDER BY ASC(?label)
'''

for row in g.query(q):
    total_patient_number_for_this_drug = int(row.responder_count) + int(row.non_responder_count)
    print(total_patient_number_for_this_drug, "patients in total took", row.label, 
          "with methotrexate co-therapy status", row.methotrexate_cotherapy,
          "of which", row.responder_count, 
          '({0:.0%})'.format(int(row.responder_count) / total_patient_number_for_this_drug), 'were responders and',
          row.non_responder_count, 
          '({0:.0%})'.format(int(row.non_responder_count) / total_patient_number_for_this_drug), 'were non-responders')
```

#### Adalimumab with methotrexate co-therapy leaves all other treatment options behind! For the sake of completeness, let's also look at the effects of methotrexate co-therapy status by itself.

```python
q = '''\
SELECT ?methotrexate_cotherapy (COUNT(DISTINCT ?responder) AS ?responder_count)(COUNT(DISTINCT ?non_responder) AS ?non_responder_count) WHERE {
    {
        ?responder ra:non-responder false . 
        ?responder ra:methotrexate-cotherapy ?methotrexate_cotherapy .

    }
    UNION 
    {
        ?non_responder ra:non-responder true .
        ?non_responder ra:methotrexate-cotherapy ?methotrexate_cotherapy .
    }
}
GROUP BY ?methotrexate_cotherapy
'''

for row in g.query(q):
    total_patient_number_for_this_drug = int(row.responder_count) + int(row.non_responder_count)
    print(total_patient_number_for_this_drug, 
          "had methotrexate status", row.methotrexate_cotherapy, 
          "of which", row.responder_count, 
          '({0:.0%})'.format(int(row.responder_count) / total_patient_number_for_this_drug), 'were responders and',
          row.non_responder_count, 
          '({0:.0%})'.format(int(row.non_responder_count) / total_patient_number_for_this_drug), 'were non-responders')
```

#### Taking methotrexate is not generally associated with a marked decrease in non-responders. Let's also look at possible age and gender discrepancies

```python
# delete triples where age is 'NA', otherwise AVG(?age) below would not work...

q = '''\


DELETE WHERE {
    ?person ra:age "NA"^^xsd:string .
}
'''

g.update(q)
```

```python
# for non-responders
q = '''\
SELECT ?label 
       ?methotrexate_cotherapy 
       (COUNT(DISTINCT ?non_responder) AS ?non_responder_count) 
       (AVG(?age) as ?average_age)
WHERE {
        ?non_responder ra:drug ?drug ;
            ra:non-responder true ;
            ra:methotrexate-cotherapy ?methotrexate_cotherapy ;
            ra:gender ?gender ;
            ra:age ?age .
        ?drug rdfs:label ?label .
} 
GROUP BY ?label ?methotrexate_cotherapy
ORDER BY ASC(?label)'''

for row in g.query(q):
    total_patient_number_for_this_drug = int(row.non_responder_count)
    print(total_patient_number_for_this_drug, "non-responders took", row.label,
          "with methotrexate status", row.methotrexate_cotherapy, 
          '-- Average age:', row.average_age)
```

```python
# for responders

q = '''\
SELECT ?label 
       ?methotrexate_cotherapy 
       (COUNT(DISTINCT ?responder) AS ?responder_count) 
       (AVG(?age) as ?average_age)
WHERE {
        ?responder ra:drug ?drug ;
            ra:non-responder false ;
            ra:methotrexate-cotherapy ?methotrexate_cotherapy ;
            ra:gender ?gender ;
            ra:age ?age .
        ?drug rdfs:label ?label .
} 
GROUP BY ?label ?methotrexate_cotherapy
ORDER BY ASC(?label)'''

for row in g.query(q):
    total_patient_number_for_this_drug = int(row.responder_count)
    print(total_patient_number_for_this_drug, "responders took", row.label,
          "with methotrexate status", row.methotrexate_cotherapy, 
          '-- Average age:', row.average_age)
```

## Simple algorithm

```python
def homebrew_predict_proba(feature_names, vectorized_data, only_genetic = False):
    fn = feature_names
    proba = list()
    
    for sample in vectorized_data:
        if only_genetic == False:
            # set initial probabilities based on drug regimen
            if (sample[fn.index('drug=adalimumab [drugbank:DB00051]')] == 1) and (sample[fn.index('methotrexate_cotherapy')] == 1):
                probability_of_non_response = 0.13
            elif (sample[fn.index('drug=adalimumab [drugbank:DB00051]')] == 1) and (sample[fn.index('methotrexate_cotherapy')] == 0):
                probability_of_non_response = 0.22
            elif (sample[fn.index('drug=etanercept [drugbank:DB00005]')] == 1) and (sample[fn.index('methotrexate_cotherapy')] == 1):
                probability_of_non_response = 0.22
            elif (sample[fn.index('drug=etanercept [drugbank:DB00005]')] == 1) and (sample[fn.index('methotrexate_cotherapy')] == 0):
                probability_of_non_response = 0.23
            elif (sample[fn.index('drug=infliximab [drugbank:DB00065]')] == 1) and (sample[fn.index('methotrexate_cotherapy')] == 1):
                probability_of_non_response = 0.27
            elif (sample[fn.index('drug=infliximab [drugbank:DB00065]')] == 1) and (sample[fn.index('methotrexate_cotherapy')] == 0):
                probability_of_non_response = 0.30
            else:
                probability_of_non_response = 0.21
            # modify probability based on gender
            if sample[fn.index('gender_male')] == 1 :
                probability_of_non_response *= 0.9
        else:
            # set initial probability
            probability_of_non_response = 0.21

        # modify probability based on pharmacogenomic markers; '1' means heterozygous, '2' means homozygous
        if sample[fn.index('http://bio2rdf.org/ra.challenge:rs548234_C')] == 2 :
            probability_of_non_response *= 0.85
        elif sample[fn.index('http://bio2rdf.org/ra.challenge:rs548234_C')] == 1 :
            probability_of_non_response *= 0.9
            
        if (sample[fn.index('methotrexate_cotherapy')] == True) or (only_genetic == True): # in case we cannot use clinical data, we ASSUME that methotrexate is given as well
            if sample[fn.index('http://bio2rdf.org/ra.challenge:rs4673993_C')] == 2:
                probability_of_non_response *= 1.3
            if sample[fn.index('http://bio2rdf.org/ra.challenge:rs7279445_T')] == 2:
                probability_of_non_response *= 1.1
            if sample[fn.index('http://bio2rdf.org/ra.challenge:rs1801394_A')] == 2:
                probability_of_non_response *= 1.15
            if sample[fn.index('http://bio2rdf.org/ra.challenge:rs13120400_C')] == 2:
                probability_of_non_response *= 1.35
            methotrexate_score = \
                sample[fn.index('http://bio2rdf.org/ra.challenge:rs7279445_T')] +\
                sample[fn.index('http://bio2rdf.org/ra.challenge:rs5760410_A')] +\
                sample[fn.index('http://bio2rdf.org/ra.challenge:rs1801394_A')] +\
                sample[fn.index('http://bio2rdf.org/ra.challenge:rs13120400_C')] 
            if methotrexate_score > 2:
                probability_of_non_response *= 1.2
 
        proba.append(round(probability_of_non_response,4))
        
    return(proba)
```

```python
from sklearn.feature_extraction import DictVectorizer

q = '''\
SELECT ?person ?identifier ?gender ?age ?drug_label ?methotrexate_cotherapy
       (GROUP_CONCAT(DISTINCT ?polymorphism_variant_homozygous; SEPARATOR = "\t") as ?polymorphism_variant_homozygous_concat)
       (GROUP_CONCAT(DISTINCT ?polymorphism_variant_heterozygous; SEPARATOR = "\t") as ?polymorphism_variant_heterozygous_concat)
WHERE {
    ?person <http://bio2rdf.org/bio2rdf_vocabulary:identifier> ?identifier ; 
        ra:gender ?gender ;
        ra:age ?age ;
        ra:methotrexate-cotherapy ?methotrexate_cotherapy  ;
        ra:drug ?drug .
        ?drug rdfs:label ?drug_label .
    FILTER NOT EXISTS { 
        ?person ra:non-responder ?non_responder .
    }
    OPTIONAL {
        ?person ra:has-two ?polymorphism_variant_homozygous .
        ?polymorphism_variant_homozygous  a ra:minor_variant ;
            a ?snp .
        FILTER(REGEX(STR(?snp), "rs7279445|rs5760410|rs1801394|rs13120400|rs4673993|rs548234"))
        #FILTER(REGEX(STR(?snp), "rs548234"))
    }
    OPTIONAL {
        ?person ra:has-one ?polymorphism_variant_heterozygous .
        ?polymorphism_variant_heterozygous a ra:minor_variant ;
            a ?snp .
        FILTER(REGEX(STR(?snp), "rs7279445|rs5760410|rs1801394|rs13120400|rs4673993|rs548234"))
        #FILTER(REGEX(STR(?snp), "rs548234"))
    }
}
GROUP BY ?person ?identifier ?gender ?age ?drug_label ?methotrexate_cotherapy 
'''

patient_ids = list()
patient_data = list()

for row in g.query(q):
    
    patient_ids.append(str(row.identifier))
    
    # clinical data
    data_dict_for_this_patient = dict()
    data_dict_for_this_patient = {'gender_male': True if (str(row.gender) == 'http://bio2rdf.org/ra.challenge_vocabulary:Male') else False,
                                  'age': float(row.age),
                                  'drug': str(row.drug_label), 
                                  'methotrexate_cotherapy': bool(row.methotrexate_cotherapy)}
    #  variant data
    data_dict_for_this_patient.update({polymorphism_variant:1 
                                       for polymorphism_variant in row.polymorphism_variant_heterozygous_concat.split('\t')})
    data_dict_for_this_patient.update({polymorphism_variant:2 
                                       for polymorphism_variant in row.polymorphism_variant_homozygous_concat.split('\t')})
    
    patient_data.append(data_dict_for_this_patient)


print(len(patient_data))
print(len(patient_ids))

vec = DictVectorizer()
patient_data_vectorized = vec.fit_transform(patient_data).toarray()
print("Features:\n")
print(vec.get_feature_names())
print("\nExample data matrices:\n")
print(patient_data_vectorized[0:10])
```
